## Data Visualization

Uni-variate Analysis

```python
# Countplot is a uni-variate plot for numerical and categorical variable.
plot = sns.countplot(data=ford_df, x='model')
plt.xticks(rotation = 90)
plt.title("Models of all cars")
plt.show()
```

* The countplot displays that Fiesta, Focus and Kuga is having the highest count of car model present in the data with 7000, 4800 and 2500, and other model of cars are very less count having below 1500.

```python
# Histplot is a uni-variate plot for numerical and categorical variable.
sns.histplot(data=ford_df, x='price', hue='transmission')
plt.title("Price of cars")
```

* The histplot of price of cars displays that the heighest count of cars price is having around 100000. There are more maual cars present in data.

* Based on this plot as the price increases to 100000 the count of cars increases and after 100000 price range the count of cars decrease with the increase in prices of cars. 

```python
# Countplot is a uni-variate plot for numerical and categorical variable.
sns.countplot(data = ford_df, x='transmission')
```

* The ford car data consists of cars with Manual transmission is having the heighest count of 15000.

* The Automatic and semi-automatic is having very less cars with the count of below 2000.

```python
# Histplot is a uni-variate plot for numerical and categorical variable.
sns.histplot(data=ford_df, x='mileage', hue='fuelType')
plt.title("Mileage driven by cars")
```

* The mileage of ford cars displays that the mileage intially for cars below 25000 is having highest count of cars available with count of 800.

* After the 25000 mileage of cars the count decreases and the heighest mileage driven cars are 175000.

* More cars are petrol more than 200 count all the cars are petrol cars and below 200 cars are remaining fueltype of cars in ford cars dataset.

```python
# Histplot is a uni-variate plot for numerical and categorical variable.
sns.histplot(data=ford_df, x='model', hue='fuelType')
plt.xticks(rotation=90)
plt.title("Model driven by cars")
```

* Histplot shows that the count of car model fueltype, some cars having the different fueltype options in car model.

```python
# Countplot is a uni-variate plot for categorical variable.
sns.countplot(data=ford_df, x='fuelType')
plt.title("fueltype of ford cars")
```

* The fuel type of petrol cars is having the highest count of 12000 and diesel with 6000 cars and Hybrid and Electric cars having the count below 500.

```python
# Countplot is a uni-variate plot for categorical variable.
sns.displot(data=ford_df, x='mpg', kind='kde')
plt.title("mpg of ford cars")
```

```python
# Boxplot is a uni-variate plot for numerical variable.
sns.boxplot(data= ford_df, x='mpg')
```

* Boxplot for mpg(miles per gallon) is having an outlier very high at 200 mpg. We need to treat this outlier values at rightside.

```python
# Boxplot is a uni-variate plot for numerical variable.
sns.boxplot(data= ford_df, x='engineSize')
```

* Boxplot for engine size is having very few values at right side of plot.

```python
# Boxplot is a uni-variate plot for numerical variable.
sns.boxplot(data= ford_df, x='price')
```

* Boxplot tells us that there i any outliers present in data or not, we can see here there are few values are having towards right side as outliers knowns as right skewness.

* For model prediction these outliers may effect the model prediction so, we need to treat these outliers with some techniques like Z-transformation or by threshold values.

```python
# Boxplot is a uni-variate plot for numerical variable.
sns.boxplot(data=ford_df, x='mileage')
```

* Here also in mileage there are outliers as values of high range of very less values present on right side so, it is a right skewed plot.

* We need threat this outliers with z-transformation.

## Bi-Variate Analysis

```python
# Scatterplot is a bi-variate plot between numerical and categorical
plt.figure(figsize=(10,5))
sns.scatterplot(data=ford_df, x='model', y='mpg', hue='fuelType')
plt.xticks(rotation=90)
plt.title(" Model vs mpg ")
plt.show()
```

* Scatterplot for model and mpg(miles per gallon) shows that petrol cars is having 75mpg and diesel is having 100mpg.

```python
# Scatterplot is a bi-variate plot between numerical and categorical
plt.figure(figsize=(10,5))
sns.scatterplot(data=ford_df, x='model', y='engineSize', hue='transmission')
plt.xticks(rotation=90)
plt.title(" Model vs engine Size ")
plt.show()
```

* Scatterplot for model and engine size shows that each model of car is having different engine sizes and most of the cars are manual.

```python
# Scatterplot is a bi-variate plot between numerical and categorical
plt.figure(figsize=(10,5))
sns.scatterplot(data=ford_df, x='model', y='price', hue='fuelType')
plt.xticks(rotation=90)
plt.title(" Model vs price ")
plt.show()
```

* scatterplot between model and price based on fueltype of
cars show that there is a close relation ship between them more cars having more than one fueltype.

* There are some cars having only petrol or diesel fueltype. In this ford mustang,KA,puma,ka+,streetka,escort and ford focus are only petrol cars. Among these cars ford mustang is having the heighest price.

* We can observe that diesel cars are having the second heighest price cars in ford.



```python
# Scatterplot is a bi-variate plot between numerical and categorical
plt.figure(figsize=(10,5))
sns.scatterplot(data=ford_df, x='model', y='price', hue='transmission')
plt.xticks(rotation=90)
plt.title(" Model vs price ")
plt.show()
```

* Scatterplot for model and price based on transmission shows that there is a close relationship between all the transmission of cars.

* Among them manual cars are more and price of manual cars is having the heighest

```python
# Scatterplot is a bi-variate plot between numerical and categorical
plt.figure(figsize=(10,5))
sns.scatterplot(data=ford_df, x='model', y='mileage', hue='transmission')
plt.xticks(rotation=90)
plt.title(" Model vs mileage ")
plt.show()
```

```python
# pointplot is a bi-variate plot between numerical and categorical
plt.figure(figsize=(10,6))
sns.pointplot(data=ford_df, x='model', y='price')
plt.xticks(rotation=90)
plt.title("model vs price pointplot")
plt.show()
```

* Pointplot tells that the peak prices of a particilar model connected dots through line of thier peak price.

```python
# lineplot is a bivariate plot between two numerical variables
plt.figure(figsize=(15,8))
sns.lineplot(data=ford_df, x='mileage', y='price')
plt.title("lineplot for Price vs mileage")
```

* The lineplot shows that the driven mileage of car increases the price of particular car decreases so, less mileage driven car having the highest price.
